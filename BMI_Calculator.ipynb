```python
def calculate_bmi(weight_kg: float, height_m: float) -> float:
    return weight_kg / (height_m ** 2)


def get_bmi_category(bmi: float) -> str:
    if bmi < 18.5:
        return "Underweight"
    elif bmi < 25.0:
        return "Normal weight"
    elif bmi < 30.0:
        return "Overweight"
    else:
        return "Obese"


def get_float_input(prompt: str) -> float:
    while True:
        try:
            value = float(input(prompt))
            if value <= 0:
                print("Please enter a positive number.")
                continue
            return value
        except ValueError:
            print("Invalid input. Please enter a number.")


def get_unit_choice() -> str:
    while True:
        choice = input("Use (M)etric (kg/cm) or (I)mperial (lbs/in)? ").strip().upper()
        if choice in ("M", "I"):
            return choice
        print("Please enter M or I.")


def main():
    print("=" * 40)
    print("         BMI CALCULATOR")
    print("=" * 40)

    unit = get_unit_choice()
    print()

    if unit == "M":
        weight_kg = get_float_input("Enter weight (kg): ")
        height_cm = get_float_input("Enter height (cm): ")
        height_m = height_cm / 100.0
    else:
        weight_lbs = get_float_input("Enter weight (lbs): ")
        height_in = get_float_input("Enter height (inches): ")
        weight_kg = weight_lbs * 0.453592
        height_m = height_in * 0.0254

    bmi = calculate_bmi(weight_kg, height_m)
    category = get_bmi_category(bmi)

    print()
    print("-" * 40)
    print(f"BMI: {bmi:.1f}")
    print(f"Category: {category}")
    print("-" * 40)
    print("BMI Categories:")
    print("< 18.5      Underweight")
    print("18.5-24.9   Normal weight")
    print("25.0-29.9   Overweight")
    print(">= 30.0     Obese")
    print("=" * 40)


if __name__ == "__main__":
    main()
```
